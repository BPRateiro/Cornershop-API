## Carregando todos os dados do diretório `./data/` em um único dataframe

```python
import glob
import pandas as pd
def read_json_files_as_dataframe(directory):
    """
    Read all JSON files from a directory and combine them into a single DataFrame.

    Args:
        directory (str): The path to the directory containing the JSON files.

    Returns:
        pd.DataFrame: The resulting DataFrame that combines all the JSON files.

    """
    file_paths = glob.glob(f"{directory}/*.json")

    dataframes = [pd.read_json(file_path) for file_path in file_paths]
    combined_dataframe = pd.concat(dataframes, ignore_index=True)
    return combined_dataframe

data_dir = 'data/'
df = read_json_files_as_dataframe(data_dir)
```

```python
def remove_duplicate_rows(df):
    """
    Remove rows from a DataFrame where all column values are exactly the same.

    Args:
        df (pd.DataFrame): The DataFrame to remove duplicate rows from.

    Returns:
        pd.DataFrame: The resulting DataFrame with duplicate rows removed.

    """
    num_duplicates = df.duplicated(keep=False, subset=df.columns.difference(['index'])).sum()
    df_without_duplicates = df.drop_duplicates(keep=False, subset=df.columns.difference(['index']))
    print(f"Removed {num_duplicates} duplicate rows")
    return df_without_duplicates

df = remove_duplicate_rows(df)
```

```python
def remove_duplicate_rows_except_search(df):
    """
    Remove rows from a DataFrame where all column values are exactly the same.

    Args:
        df (pd.DataFrame): The DataFrame to remove duplicate rows from.

    Returns:
        pd.DataFrame: The resulting DataFrame with duplicate rows removed.

    """
    num_duplicates = df.duplicated(keep=False, subset=df.columns.difference(['search_term'])).sum()
    df_without_duplicates = df.drop_duplicates(subset=df.columns.difference(['search_term']))
    print(f"Removed {num_duplicates} duplicate rows")
    return df_without_duplicates

df = remove_duplicate_rows_except_search(df)
```
