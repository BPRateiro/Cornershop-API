## Scripts

```python
def print_results(results):
    for data in results:
        store_name = data['store']['name']
        store_description = data['store']['description']
        search_term = data['search_result']['search_term']
        aisles = data['search_result']['aisles']

        print(f"Nome da loja: {store_name}")
        print(f"Descrição da loja: {store_description}")
        print(f"Termo de busca: {search_term}")
        print("Aisles:")
        for aisle in aisles:
            aisle_name = aisle['aisle_name']
            print(f"  - Nome do corredor: {aisle_name}")
            products = aisle['products']
            for product in products:
                product_name = product['name']
                product_price = product['pricing']['price']['amount']
                print(f"    * Produto: {product_name} - Preço: R$ {product_price}")
            print()
        print()

print_results(results)
```
